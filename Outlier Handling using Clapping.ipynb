```python
Q1=sp["ApplicantIncome"].quantile(0.25)  #checking the percentages
Q3=sp["ApplicantIncome"].quantile(0.75)
IQR=Q3-Q1         #finding the IQR
LL=Q1-1.5*IQR     #finding the lower limit
UL=Q3+1.5*IQR     #finding the upper limit
```

```python
(Q1,Q3,IQR,LL,UL)  #all values
```

```python
sp[sp["ApplicantIncome"]<LL]    #checking the values lesser than lowerlimit
```

```python
sp[sp["ApplicantIncome"]>UL]     #checking the values greater than upperlimit
```

```python
sp.loc[sp['ApplicantIncome'] > UL, 'ApplicantIncome'] = UL   #changing the outliers to upperlimit
```
